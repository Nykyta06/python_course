Реалізуйте структуру даних, що є розширеною структурою стек. Необхідно підтримувати додавання елемента на вершину стека, видалення з вершини стека, і необхідно підтримувати операції складання, віднімання, множення та цілісного поділу.

Операція додавання на стеку визначається так. Зі стека знімається верхній елемент (top1), потім знімається наступний верхній елемент (top2), і потім як результат операції складання на вершину стека кладеться елемент, рівний top1 + top2.

Аналогічним чином визначаються операції віднімання (top1 - top2), множення (top1 * top2) і цілісного поділу (top1 // top2).

*Примітка*

Для додавання елемента на стек використовується метод append, а зняття зі стека – метод pop.
Гарантується, що операції будуть здійснюватися тільки коли в стеку є хоча б два елементи.

````python
# Код  написати тут
class ExtendedStack(list):

  def sum(self):
    if len(self)>1:
      self.append (self.pop() + self.pop())

  def sub(self):
    if len(self)>1:
      self.append (self.pop() - self.pop())

  def mul(self):
    if len(self)>1:
      self.append (self.pop() * self.pop())

  def div(self):
    if len(self)>1:
      self.append (self.pop() // self.pop())

#Перевірка коду класу

def test():
    ex_stack = ExtendedStack([1, 2, 3, 4, -3, 3, 5, 10])
    ex_stack.div()
    assert ex_stack.pop() == 2
    ex_stack.sub()
    assert ex_stack.pop() == 6
    ex_stack.sum()
    assert ex_stack.pop() == 7
    ex_stack.mul()
    assert ex_stack.pop() == 2
    assert len(ex_stack) == 0

test()
````

Розглянемо клас Loggable:
```python
import time

class Loggable:
     def log(self, msg):
         print(str(time.ctime()) + ": " + str(msg))
```
У нього є рівно один метод log, який дозволяє виводити в балку (в даному випадку в stdout) якесь повідомлення, додаючи при цьому поточний час.

Реалізуйте клас LoggableList, успадкувавши його від класів list і Loggable таким чином, щоб при додаванні елемента до списку за допомогою методу append в лог надсилалося повідомлення, що складається з щойно доданого елемента.


*Примітка*

Ваша програма не повинна містити клас Loggable.

````python
# Код  написати тут

import time

class Loggable:
     def log(self, msg):
         print(str(time.ctime()) + ": " + str(msg))

class LoggableList(Loggable, list):
    def append(self, element):
        self.log(element)
        super().append(element)

test = LoggableList([1,2,3])
test.append(4)
````
